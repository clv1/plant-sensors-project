```python
import pandas as pd
import altair as alt
```

```python
plants_data = pd.concat(
    map(pd.read_csv, ['example_dataset_1.csv', 'example_dataset_2.csv', 'example_dataset_3.csv',
                      'example_dataset_4.csv', 'example_dataset_5.csv', 'example_dataset_6.csv']), ignore_index=True)
```

```python
plants_data
```

```python
plants_data['last_watered'] = pd.to_datetime(plants_data['last_watered'])
plants_data['recording_taken'] = pd.to_datetime(plants_data['recording_taken'])
```

```python
plants_data_sorted_by_last_watered = plants_data.sort_values(by='last_watered')
plants_data_sorted_by_last_watered
```

```python
plants_sorted_by_recording_taken = plants_data.sort_values(by='recording_taken')
plants_sorted_by_recording_taken 
```

```python
alt.Chart(plants_data).mark_line().encode(
    x='recording_taken:T',
    y='soil_moisture:Q',
    color = 'name:N'
)
```

```python
alt.Chart(plants_data).mark_circle().encode(
    x='recording_taken:T',
    y='last_watered:T',
    color = 'name:N'
)
```

```python
plant_19 = plants_data[plants_data['plant_id'] == 19]
```

```python
alt.Chart(plant_19).mark_line().encode(
    x='recording_taken:T',
    y='temperature:Q',
    color = 'name:N'
)
```
